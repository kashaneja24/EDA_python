```python
df = df.drop(['1.4','3.5'], axis=1)
df.head(5)
```

```python
df = df.rename(columns={"0.2": "00.2", "1.4": "01.4"})
df.head(5)
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
 df = df.drop_duplicates()
df.head()
```

```python
df.count()
```

```python
print(df.isnull().sum())
```
